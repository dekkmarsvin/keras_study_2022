```python
norm_df = raw_data.drop(["Surname"], axis=1)
norm_df["Geography"] = pd.Categorical(norm_df["Geography"])
norm_df["Geography"] = norm_df.Geography.cat.codes
norm_df["Gender"] = pd.Categorical(norm_df["Gender"])
norm_df["Gender"] = norm_df.Gender.cat.codes

target = norm_df.pop("Exited")
dataset = tf.data.Dataset.from_tensor_slices((norm_df.values, target.values))
```

```python
train_dataset = dataset.shuffle(len(norm_df)).batch(1)
```
